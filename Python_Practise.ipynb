## 2. OOPs
```
# @staticmethod
  - Utility/helper class
# @classmethod
  - factory method
```


````python
## Encapsulation - Bundle data + methods, restrict access

class BankAccount:
  def __init__(self, balance):
    self.__balance = balance ##Private

  def deposit(self, amount):
    self.__balance += amount

  def get_balance(self):
    return self.__balance
````

````python
b = BankAccount(100)
````

````python
b.deposit(200)
````

````python
b.get_balance()
````

````python
from abc import ABC, abstractmethod
## Abstraction - hide complexity, show only essentials
##
class Shape(ABC):
  @abstractmethod
  def area(self):
    pass
````

````python
class MyClass:
  count = 55

  @staticmethod
  def add(a, b):
    return a + b

  @classmethod
  def get_count(cls):
    return cls.count  ##class level data or create instances(factory pattern)
````

````python
print(MyClass.add(2, 3))
print(MyClass.get_count())
````

````python
class TEST:
  count = 20

  @staticmethod
  def count_(a, b):
    return a + b

  @classmethod
  def get_count(cls):
    return cls.count

````

````python
k = TEST()

print(k.count_(2,3))

print(TEST.get_count())
````

## 3. Palindrome check

````python
def palindrome_check(s):
  left, right = 0, len(s) - 1
  while left < right:
    if s[left] != s[right]:
      return False
    left += 1
    right -= 1

  return True
````

````python
def palindrome_check(s):
  left, right = 0, len(s) - 1

  while left < right:

    if s[left] != s[right]:
      return False

    left += 1
    right -= 1

  return True
````

````python
palindrome_check('aabkbaa')
````

````python
palindrome_check('racecar')
````

## 4. Inheritance

````python
class Animal:
  def speak(self):
    return "Not Defined"

class Dog(Animal):
  def speak(self):
    return "Woof" # Overrides parent

class Cat(Animal):
  pass          # Inherit parent class
````

````python
d = Dog()
c = Cat()

print(d.speak(), ":::", c.speak())
````

## 5. Frequeny count (Imp for TF-IDF/BM25)

````python
def frequency_count(lst):
  dict_ = {}

  for l in lst:
    if l in dict_:
      dict_[l] += 1
    else:
      dict_[l] = 1

  return dict_
````

````python
frequency_count([1,1,4,8,4,3,1])
````

````python
text_ = '''Some corpora have further structured levels of analysis applied. In particular, smaller corpora may be fully parsed. Such corpora are usually called Treebanks or Parsed Corpora. The difficulty of ensuring that the entire corpus is completely and consistently annotated means that these corpora are usually smaller, containing around one to three million words. '''
````

````python
chunking_with_overlap_(text_)
````

````python
def chunking_with_overlap(text_):

  words = text_.split()

  ## Chunk size 10 and overlap 2
  dict_chunks = {}
  start = 1

  for i in range(0, len(words), 7):
    dict_chunks[start] = (words[i:i+9])
    start += 1

  return dict_chunks
````

````python
print(chunking_with_overlap(text_))
````

## 8. Decorator
```Function that wraps another function to extent or modify its behavior```


````python
import time
import functools

def timer(func):
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        start_time = time.perf_counter()
        result = func(*args, **kwargs)
        end_time = time.perf_counter()

        print(f"Executed {func.__name__} in {end_time - start_time:.4f}s")
        return result
    return wrapper

@timer
def slow_function():
    time.sleep(2)
    return 'done'

print(slow_function())
````

## 10. Difference between sort and sorted.

  - The key difference is that `list.sort()` sorts a list **in-place** (modifying the original) and returns None, while `sorted(iterable)` returns a new sorted list leaving the original unchanged.
  - `sort()` is a list method, whereas `sorted()` works on any iterable (strings, tuples, dicts).

````python
list_ = [5,3,9,1]

print(list_.sort())
````

````python
list_2 = [5,3,9,1]
sorted(list_2)
````
