| :--- | :--- | :--- | :--- |
| `map` | Transforms every item | Sequence | Square every number |
| `filter` | Keeps items meeting a condition | Sequence | Only even numbers |
| `reduce` | Combines all items | Single Value | Multiply all numbers together |

## ENUMS

Enums solve the "magic string/number" problem. When you write status == "shipped", a typo silently passes. When you write status == OrderStatus.SHIPPED, Python tells you immediately if the member doesn't exist. They're iterable, hashable (usable as dict keys), and the auto() utility auto-assigns values so you don't have to manage numbers yourself. IntEnum is useful when you need to compare against raw integers (like HTTP status codes). You'll see Enums heavily used in FastAPI models and LangGraph state machines.

````python
from enum import Enum

class UploadStatus(Enum):
  UPLOADING = 'uploading'
  SUCCESS = 'success'


status = UploadStatus.SUCCESS

print(status, status.value)
````

## 3. Palindrome check

````python
def palindrome_check(s):
  left, right = 0, len(s) - 1
  while left < right:
    if s[left] != s[right]:
      return False
    left += 1
    right -= 1

  return True
````

````python
def palindrome_check(s):
  left, right = 0, len(s) - 1

  while left < right:

    if s[left] != s[right]:
      return False

    left += 1
    right -= 1

  return True
````

````python
palindrome_check('aabkbaa')
````

## 5. Frequeny count (Imp for TF-IDF/BM25)

````python
def frequency_count(lst):
  dict_ = {}

  for l in lst:
    if l in dict_:
      dict_[l] += 1
    else:
      dict_[l] = 1

  return dict_
````

````python
def chunking_with_overlap(text_):

  words = text_.split()

  ## Chunk size 10 and overlap 2
  dict_chunks = {}
  start = 1

  for i in range(0, len(words), 7):
    dict_chunks[start] = (words[i:i+9])
    start += 1

  return dict_chunks
````

````python
## Non Local ==> Commonly used in closures to maintain state without using global variables.
````

````python
def outer():
    count = 0
    def inner():
        nonlocal count  # Targets 'count' in outer() --> Get count value from outer loop
        count += 1
        return count
    return inner
````

## Why it Matters
   - Safety: Using local variables is generally safer because it prevents "side effects" where one part of your code accidentally breaks another.
  - Encapsulation: Nonlocal allows you to hide data within a function while still letting inner parts of that function update it.
  - Shared State: Global is best for constants or configuration that every part of your program needs to know.

## 8. Decorator
```Function that wraps another function to extent or modify its behavior```


````python
import time
import functools

def timer(func):
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        start_time = time.perf_counter()
        result = func(*args, **kwargs)
        end_time = time.perf_counter()

        print(f"Executed {func.__name__} in {end_time - start_time:.4f}s")
        return result
    return wrapper

@timer
def slow_function():
    time.sleep(2)
    return 'done'

print(slow_function())
````

## LRU Cache:
An `LRU (Least Recently Used)` Cache is a specialized data structure that stores a limited amount of information, automatically removing the item that hasn't been accessed for the longest time when it reaches capacity. It operates on the principle of temporal locality, assuming that data you used recently is likely to be needed again soon.

## **GIL**:
  - The `Global Interpreter Lock (GIL)` is a specialized "mutex" (mutual exclusion lock) used in CPython, the default and most popular implementation of Python. It ensures that only one thread can execute Python bytecode at a time, even on systems with multiple CPU cores.
  - Native Extensions: Libraries like `NumPy` or `TensorFlow` often release the GIL when performing heavy computations in C, allowing them to bypass the lock and run in parallel.

## Purpose and Function: **name-main idiom**

  - When run as a script: The Python interpreter automatically sets the special __name__ variable to the string "__main__" for the file that is executed as the main program. The code within the if block runs in this scenario.
  - When imported as a module: If the file is imported into another script, the __name__ variable is set to the module's name (e.g., if you import a file named my_module.py, __name__ becomes "my_module"). The condition __name__ == "__main__" evaluates to False, and the code inside the if block is skipped.
