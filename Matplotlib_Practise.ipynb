```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

# 📊 Basic Matplotlib Functions


#### `plt.figure()`
Creates a new figure or canvas where the plot will be drawn.

---

#### `plt.plot()` - Basic **line plot**.

#### `plt.title()` - Adds a **title** to the plot.

#### `plt.xlabel()` - Adds a label to the **X-axis**.

#### `plt.ylabel()` - Adds a label to the **Y-axis**.

#### `plt.bar()` - Used to create a **bar chart**.

#### `plt.show()` - Displays the final plot on the screen.


```python
df = pd.read_csv('bmw_global_sales_2018_2025.csv')
```

```python
df.head(2)
```

```python
df.describe()
```

```python
df.columns
```

```python
## Trend Analysis

plt.bar(df['Year'], df['GDP_Growth'])
plt.xlabel('Year')
plt.ylabel('GDP Growth')
plt.title('GCP Growth YoY')
plt.show()
```

```python
trend_sales = df.groupby('Year')['Units_Sold'].sum().reset_index()
```

```python
plt.plot(trend_sales['Year'], trend_sales['Units_Sold'])
plt.ylabel('Sales in Millions')
plt.xlabel('Year')
plt.show()
```

```python
trend_sales_1 = df.groupby('Year')['Revenue_EUR'].sum().reset_index()
```

```python
plt.plot(trend_sales_1['Year'], trend_sales_1['Revenue_EUR'])
plt.xlabel('Year')
plt.ylabel('Revenue in Euro')
plt.show()
```

```python
plt.plot(trend_sales['Year'], trend_sales['Revenue_EUR'])
plt.ylabel('Sales in Millions')
plt.xlabel('Year')
plt.show()
```
