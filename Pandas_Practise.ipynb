### 1. **IMP** Concept - **Quick Comparison: `merge`, `join`, and `concat`**


| Feature | `pd.merge()` | `df.join()` | `pd.concat()` |
| :--- | :--- | :--- | :--- |
| **Logic** | **SQL-style** (Column-on-Column) | **Index-based** | 	Stacking along an axis |
| **Default Join** | `inner` | `left` | `outer` |
| **Primary Axis** | Horizontal (Columns) | Horizontal (Columns) | Vertical (Rows) or Horizontal |
| **Best For** | Relational data with shared IDs | Lookups where index is the key | Combining multiple similar CSVs |

# 🐼 Pandas Essentials Cheat Sheet

Use these methods to get a "feel" for your dataset before cleaning it.

*   **`df.shape`**: Returns a tuple `(rows, columns)` representing the size of the data.
*   **`df.columns` / `df.keys()`**: List all column names in the DataFrame.

---

*   **`df.isnull().sum()`**

*   **`df.dropna`**
*   **`df.fillna`**

# 📊 Pandas: Grouping & Sorting

### 🏗️ 1. Sorting Data
Rearrange your rows based on column values.


| Method | Description |
| :--- | :--- |
| `df.sort_values('col')` | Sort by a column (default: **Ascending**). |
| `df.sort_values('col', ascending=False)` | Sort by a column (Descending). |
| `df.sort_index()` | Sort rows by their **Index** labels. |
| `df.nlargest(5, 'col')` | Quick way to get the **Top 5** highest values. |

---

### 🧱 2. Grouping Data (`groupby`)
The **Split-Apply-Combine** strategy. Use this to calculate metrics per category.

*   **`df.groupby('category').mean()`**: Get averages for every numerical column per group.
*   **`df.groupby('category')['sales'].sum()`**: Sum up a specific column (`sales`) per group.
*   **`df.groupby('category').size()`**: Count the number of rows in each group.
*   **`df[col_name].value_counts()`**: .

---

### 🧪 3. Advanced Aggregation (`agg`)
Use `.agg()` when you need **multiple** statistics at once.

```python
# Get Min, Max, and Mean for the 'Price' column
df.groupby('City')['Price'].agg(['min', 'max', 'mean'])

# Different stats for different columns
df.groupby('Dept').agg({'Salary': 'mean', 'ID': 'count'})

# 🎯 Pandas: Selecting & Slicing Data

### 📍 1. `loc` vs `iloc`
The two primary ways to "grab" data. Think of **loc** as *location by name* and **iloc** as *integer-location*.


| Method | Selection Type | Logic | Syntax Example |
| :--- | :--- | :--- | :--- |
| **`.loc[]`** | **Label-based** | Uses Row/Column **Names** | `df.loc[0:5, 'Name']` |
| **`.iloc[]`** | **Position-based** | Uses Row/Column **Index (0,1,2...)** | `df.iloc[0:5, 0:2]` |

---

### 🎨 2. Fancy Indexing
Passing a **list** of specific labels or positions to extract exactly what you need.

*   **Specific Rows/Cols**: `df.loc[[1, 5, 10], ['Name', 'Salary']]`
*   **Reordering**: You can use fancy indexing to reorder columns on the fly:  
    `df[['Salary', 'Name', 'ID']]`

---

### 🎭 3. Boolean Masking (Filtering)
Selecting rows based on a condition (the "Excel Filter" equivalent).

```python
# 1. Create a mask (True/False list)
mask = df['Age'] > 25

# 2. Apply it to the dataframe
df[mask]

# Combined shortcut:
df[df['City'] == 'London']

# ✂️ Pandas: Column Management & Duplicates

### 🏗️ 1. Modifying Columns
Change the structure of your DataFrame by removing or renaming fields.


| Operation | Syntax Example | Key Argument |
| :--- | :--- | :--- |
| **Drop Columns** | `df.drop(columns=['Age', 'ID'])` | `inplace=True` to modify original |
| **Drop Rows** | `df.drop(index=[0, 1, 2])` | Removes rows by index labels |
| **Rename** | `df.rename(columns={'old':'new'})` | Takes a dictionary `{old: new}` |
| **Assign** | `df.assign(total=df['a'] + df['b'])` | Create new columns functionally |

---

### 👯 2. Handling Duplicates
Find and remove redundant rows to ensure data integrity.

*   **`df.duplicated()`**: Returns a Boolean Series (`True` if the row is a repeat).
*   **`df.duplicated().sum()`**: Quick way to count how many rows are exact matches.
*   **`df.drop_duplicates()`**: Removes all identical rows, keeping the **first** one by default.

---

### 🧪 3. Advanced Cleanup
Sometimes you only care about duplicates in **specific** columns.

```python
# Check for duplicate IDs only
df.duplicated(subset=['emp_id'])

# Keep the LAST occurrence instead of the first
df.drop_duplicates(subset=['email'], keep='last')

# Rename all columns to lowercase & replace spaces with underscores
df.columns = df.columns.str.lower().str.replace(' ', '_')

# 🔄 Pandas: Reshaping with Melt & Pivot

### 🧊 1. Wide vs. Long Data
*   **Wide Data**: Multiple columns for different categories (e.g., `Jan`, `Feb`, `Mar`). Great for human reading.
*   **Long Data**: One column for "Date" and one for "Value". Great for analysis and plotting.

---

### 🫠 2. `pd.melt()`: Turning Wide to Long
Use this to "unpivot" your data. It turns columns into rows.


| Argument | Description |
| :--- | :--- |
| **`id_vars`** | Columns you want to **keep** (e.g., 'Name'). |
| **`value_vars`** | Columns you want to **unstack** (e.g., 'Jan', 'Feb'). |
| **`var_name`** | New name for the unstacked column header. |
| **`value_name`** | New name for the data values. |

```python
# Example: Convert monthly sales columns into one 'Month' column
df_long = pd.melt(df, id_vars=['Product'], var_name='Month', value_name='Sales')
```

### 📐 3. `df.pivot()`: Turning Long to Wide
The opposite of melt. It turns unique row values into column headers.
index: What stays as your rows (e.g., 'Date').
columns: What becomes your new column headers (e.g., 'City').
values: What fills the cells.

```python
df_wide = df.pivot(index='Date', columns='City', values='Temperature')
```

# 🛠️ Pandas: Datatype Conversion & Parquet Files

### 🔄 1. Datatype Conversion
Converting columns to the correct type is essential for saving memory and enabling specific functions (like date math).


| Method | Best For | Key Feature |
| :--- | :--- | :--- |
| **`.astype()`** | Simple, forced casting | `df['col'].astype('int64')` |
| **`pd.to_numeric()`** | Messy numbers/strings | Use `errors='coerce'` to turn junk into `NaN` |
| **`pd.to_datetime()`** | Dates and Timestamps | Intelligent parsing of multiple date formats |
| **`.astype('category')`** | Low-cardinality strings | Drastically reduces memory for repeated text |

---

### 📦 2. Creating Parquet Files
**Parquet** is a columnar storage format that is much faster and smaller than CSV. It also "remembers" your datatypes!

# Read from Parquet
