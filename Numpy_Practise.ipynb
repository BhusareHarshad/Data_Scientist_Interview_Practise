# NumPy Basics: Quick Reference

### 1. Python List vs. NumPy Array
NumPy arrays are the backbone of data science because they are faster and more memory-efficient than standard Python lists.


| Feature | **Python Lists** | **NumPy Arrays** |
| :--- | :--- | :--- |
| **Data Type** | Heterogeneous (mixed types) | **Homogeneous** (floats or ints or str) |
| **Memory** | High (stores object references) | **Low** (compact, contiguous blocks) |
| **Performance** | Slow (requires loops) | **Fast** (vectorized operations) |
| **Best For** | General purpose collections | Mathematical & scientific computing |

---
### **2. Cheat Sheet Syntax: Array Creation**


| Method | Purpose | Example Syntax |
| :--- | :--- | :--- |
| `np.array()` | Convert a Python list to a NumPy array | `np.array([1, 2, 3])` |
| `np.arange()` | Create a range of values (start, stop, step) | `np.arange(0, 10, 2)` |
| `np.zeros()` | Create an array filled with 0.0 (floats) | `np.zeros((3, 4))` |
| `np.ones()` | Create an array filled with 1.0 (floats) | `np.ones((3, 4))` |


---

### 3. Dimensions and Shape
Understanding the structure of your data is key to avoiding errors during matrix operations.

*   **`.ndim`**: The number of axes (dimensions). 1D = Vector, 2D = Matrix, 3D = Tensor.
*   **`.shape`**: A tuple representing the size of the array along each axis (e.g., `(rows, columns)`).

# **NumPy: Math, Sorting & Aggregation**

### **1. Mathematical Operations**
NumPy performs **element-wise** operations by default, which is much faster than Python loops.

Efficiently organizing data or finding specific values.


| Operation | Method | Description |

# **NumPy Broadcasting**

Broadcasting allows NumPy to perform arithmetic operations on arrays of **different shapes**, as long as they are compatible.

---

### **1. Rules of Broadcasting**
For two arrays to be compatible for broadcasting, NumPy compares their dimensions from **right to left**:
1.  The dimensions must be **equal**, OR
2.  One of the dimensions must be **1**.


| Scenario | Array A Shape | Array B Shape | Resulting Shape |
| :--- | :--- | :--- | :--- |
| **Scalar + 1D** | `(3,)` | `(1,)` or Scalar | `(3,)` |
| **2D + 1D** | `(3, 3)` | `(3,)` | `(3, 3)` |
| **Compatible** | `(4, 3)` | `(1, 3)` | `(4, 3)` |
| **Incompatible**| `(4, 3)` | `(4, 2)` | **Error!** |

---

### **2. Cheat Sheet Syntax**


| **Operation** | **Purpose** | **Example Syntax** |
| :--- | :--- | :--- |
| **Scalar Broadcast** | Add/Mult a single number to every element | `arr + 5` |
| **Row-wise** | Apply a 1D vector to every row of a 2D matrix | `matrix + row_vec` |
| **Column-wise** | Apply a 1D vector to every column (needs reshape) | `matrix + col_vec.reshape(-1, 1)` |

---

### **Quick Start Code**

```python
import numpy as np

# 1. Scalar Broadcasting
a = np.array([10, 20, 30])
print(a + 5)             # Output: [15, 25, 35]

# 2. 2D + 1D Broadcasting
matrix = np.array([[1, 2, 3],
                   [4, 5, 6]]) # Shape (2, 3)
row_vec = np.array([1, 0, 1])  # Shape (3,)

# The row_vec is "stretched" to match the 2 rows of the matrix
print(matrix + row_vec)  
# Result: [[2, 2, 4], [5, 5, 7]]

# 3. Handling Incompatible Shapes
# To broadcast column-wise, the vector must have a shape of (2, 1)
col_vec = np.array([10, 20])   # Shape (2,) -> Needs reshape
print(matrix + col_vec.reshape(2, 1))

# **NumPy Array Splitting**

Splitting is the opposite of stacking; it breaks one array into multiple smaller sub-arrays.

---

### **1. Split Operations**


| **Operation** | **Method** | **Description** | **Requirement** |
| :--- | :--- | :--- | :--- |
| **Horizontal Split** | `np.hsplit(arr, n)` | Splits array **column-wise** (horizontally). | `n` must divide columns evenly. |
| **Vertical Split** | `np.vsplit(arr, n)` | Splits array **row-wise** (vertically). | `n` must divide rows evenly. |
| **Generic Split** | `np.split(arr, n, axis)` | Splits along a specified axis (0 for rows, 1 for cols). | Flexible for N-dimensions. |

---

### **2. Cheat Sheet Syntax**


| **Function** | **Purpose** | **Example Syntax** |
| :--- | :--- | :--- |
| `np.hsplit()` | Break a matrix into multiple vertical sections | `np.hsplit(matrix, 2)` |
| `np.vsplit()` | Break a matrix into multiple horizontal sections | `np.vsplit(matrix, 3)` |
| `np.array_split()` | Split into **unequal** parts (doesn't raise error) | `np.array_split(arr, 3)` |

---

### **Quick Start Code**

```python
import numpy as np

# Create a 4x4 matrix
data = np.arange(16).reshape(4, 4)

# 1. Vertical Split (Split into 2 arrays of 2 rows each)
upper, lower = np.vsplit(data, 2)
# upper shape: (2, 4), lower shape: (2, 4)

# 2. Horizontal Split (Split into 2 arrays of 2 columns each)
left, right = np.hsplit(data, 2)
# left shape: (4, 2), right shape: (4, 2)

# 3. Using np.split (Specific axis)
# Same as vsplit
parts = np.split(data, 2, axis=0)

# 4. Unequal Splitting
# If you try to split 4 columns into 3 using hsplit, it crashes.
# np.array_split handles it by making sections as equal as possible.
unequal_parts = np.array_split(data, 3, axis=1)

# **NumPy: Linear Algebra (linalg)**

The `numpy.linalg` module provides a comprehensive suite of functions for matrix operations, solving linear systems, and finding decompositions.

---

### **1. Matrix & Vector Products**
These operations go beyond element-wise multiplication to perform standard algebraic products.


| **Operation** | **Method** | **Example Syntax** | **Description** |
| :--- | :--- | :--- | :--- |
| **Dot Product** | `np.dot(a, b)` | `a.dot(b)` | 1D: Inner product; 2D: Matrix multiplication. |
| **Matrix Multiply** | `np.matmul(a, b)` | `a @ b` | Preferred for matrix multiplication; handles 3D stacks. |
| **Inner Product** | `np.inner(a, b)` | `np.inner(a, b)` | Sum of products of corresponding elements. |
| **Outer Product** | `np.outer(a, b)` | `np.outer(a, b)` | Computes the outer product of two vectors. |

---

### **2. Solving & Inverting**
Tools for finding solutions to linear equations ($Ax = b$) and calculating matrix properties.


| **Operation** | **Method** | **Example Syntax** | **Description** |
| :--- | :--- | :--- | :--- |
| **Solve System** | `la.solve(A, b)` | `np.linalg.solve(A, b)` | Finds $x$ such that $Ax = b$. |
| **Inverse** | `la.inv(A)` | `np.linalg.inv(A)` | Computes the multiplicative inverse of a square matrix. |
| **Determinant** | `la.det(A)` | `np.linalg.det(A)` | Returns the determinant of a matrix. |
| **Matrix Rank** | `la.matrix_rank()` | `np.linalg.matrix_rank(A)`| Returns the number of linearly independent rows/cols. |
| **Trace** | `np.trace(A)` | `np.trace(A)` | Returns the sum of the diagonal elements. |

---

### **3. Decompositions & Norms**
Advanced functions often used in Machine Learning and data transformation.


| **Operation** | **Method** | **Example Syntax** | **Description** |
| :--- | :--- | :--- | :--- |
| **Eigenvalues** | `la.eig(A)` | `vals, vecs = np.linalg.eig(A)` | Returns eigenvalues and eigenvectors. |
| **SVD** | `la.svd(A)` | `u, s, vh = np.linalg.svd(A)` | Singular Value Decomposition. |
| **Norm** | `la.norm(x)` | `np.linalg.norm(x)` | Matrix or vector norm (e.g., L2 norm). |

---

### **Quick Start Code**

```python
import numpy as np

# Define a 2x2 matrix and a vector
A = np.array([[3, 1], [1, 2]])
b = np.array([9, 8])

# 1. Solving a system of linear equations (3x + y = 9; x + 2y = 8)
x = np.linalg.solve(A, b)
print(f"Solution x: {x}") # Output: [2. 3.]

# 2. Matrix multiplication (@ is shorthand for matmul)
B = np.array([[1, 0], [0, 1]]) # Identity matrix
product = A @ B

# 3. Eigenvalues and Eigenvectors
values, vectors = np.linalg.eig(A)

# 4. Transpose (Flip rows and columns)
print(A.T)
