```python
def z_stat_prop(p_observered, p,n):
  import math
  return (p_observered-p)/math.sqrt(p*(1-p)/n)
```

```python
z_stat_prop(0.6, 0.5, 100)
```

```python
2 * (1-norm.cdf(abs(z_stat_prop(0.6, 0.5, 100))))
```

```python
from scipy.stats import norm
def two_sample_z_stat_prop(p_not, p_1,p_2,n_1, n_2):
  import math
  return (p_1-p_2)/math.sqrt(p_not*(1-p_not)*(1/n_1 + 1/n_2))
```

```python
z_st = two_sample_z_stat_prop(0.49, 0.5, 0.48, 150, 300)
print(z_st)
```

```python
2*(1-norm.cdf(abs(z_st)))
```

```python
def t_stat_calculation_right_tailed(data, mean, samples):
  import numpy, math
  from scipy.stats import t

  data = numpy.array(data)

  t_stat = (data.mean() - mean)*math.sqrt(samples)/data.std()

  return round(float(t_stat), 3), round(float(1-t.cdf(t_stat, samples-1)), 3)
```
